```python
import rasterio
import rasterio.plot
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import Aerial Raster
raster_path = "data/batch_1/aerial.tif"
src = rasterio.open(raster_path)
print(src)

# Import Labels (vector)
vector_path = "data/batch_1/labels.geojson"
labelled_polygons = gpd.read_file(vector_path)
labelled_polygons = labelled_polygons.to_crs(src.crs)
```

```python
src.crs
src.res
src.transform
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
rasterio.plot.show(src, ax=ax)
labelled_polygons.plot(ax=ax, edgecolor="red", facecolor="none", lw=2)
```
